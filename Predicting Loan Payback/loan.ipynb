```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_tr = pd.read_csv("data/train.csv").drop(columns = "id")
df_ts = pd.read_csv("data/test.csv").drop(columns = "id")

df_ts
```

```python
df_tr.info()
```

```python
df_tr["loan_paid_back"].value_counts()
```

```python
df_tr['grade'] = df_tr['grade_subgrade'].str[0]
df_tr['subgrade'] = df_tr['grade_subgrade'].str[1].astype(int)

df_ts['grade'] = df_ts['grade_subgrade'].str[0]
df_ts['subgrade'] = df_ts['grade_subgrade'].str[1].astype(int)
```

```python
print(df_tr[['grade_subgrade', 'grade', 'subgrade']].head())
```

```python
df_tr = df_tr.drop(columns=["grade_subgrade"])
df_ts = df_ts.drop(columns=["grade_subgrade"])
```

```python
# map some shit with binary

df_ts["gender"] = df_ts["gender"].map({"Male": 1, "Female": 0, "Other": -1})
df_tr["gender"] = df_tr["gender"].map({"Male": 1, "Female": 0, "Other": -1})

df_ts["marital_status"] = df_ts["marital_status"].map({"Married": 1, "Single": 0, "Widowed": 0, "Divorced": -1})
df_tr["marital_status"] = df_tr["marital_status"].map({"Married": 1, "Single": 0, "Widowed": 0, "Divorced": -1})



order = ["Other", "High School", "Bachelor's", "Master's", "PhD"]
df_tr["education_level"] = (
    pd.Categorical(df_tr["education_level"], categories=order, ordered=True)
    .codes
)
df_ts["education_level"] = (
    pd.Categorical(df_ts["education_level"], categories=order, ordered=True)
    .codes
)



order = ['G', 'F', 'E', 'D', 'C', 'B', 'A']
df_tr["grade"] = (
    pd.Categorical(df_tr["grade"], categories=order, ordered=True)
    .codes
)
df_ts["grade"] = (
    pd.Categorical(df_ts["grade"], categories=order, ordered=True)
    .codes
)



```

```python
#get dummy of some categorical shit

dummies = pd.get_dummies(df_tr["employment_status"], dtype=int)
df_tr = df_tr.drop(columns=["employment_status"]).join(dummies)
dummies = pd.get_dummies(df_ts["employment_status"], dtype=int)
df_ts = df_ts.drop(columns=["employment_status"]).join(dummies)



dummies = pd.get_dummies(df_tr["loan_purpose"], dtype=int)
df_tr = df_tr.drop(columns=["loan_purpose"]).join(dummies)
dummies = pd.get_dummies(df_ts["loan_purpose"], dtype=int)
df_ts = df_ts.drop(columns=["loan_purpose"]).join(dummies)


df_tr
```

```python
df_tr['loan_to_income_ratio'] = df_tr['loan_amount'] / df_tr['annual_income']
df_tr['disposable_income'] = df_tr['annual_income'] * (1 - df_tr['debt_to_income_ratio'])
df_tr['monthly_payment_capacity'] = (df_tr['annual_income'] / 12) * (1 - df_tr['debt_to_income_ratio']) * 0.3
df_tr['total_interest_cost'] = df_tr['loan_amount'] * df_tr['interest_rate'] / 100
df_tr['interest_to_principal_ratio'] = df_tr['total_interest_cost'] / df_tr['loan_amount']
df_tr['loan_burden_index'] = (df_tr['loan_amount'] * df_tr['interest_rate'] / 100) / df_tr['annual_income']
```

```python
df_ts['loan_to_income_ratio'] = df_ts['loan_amount'] / df_ts['annual_income']
df_ts['disposable_income'] = df_ts['annual_income'] * (1 - df_ts['debt_to_income_ratio'])
df_ts['monthly_payment_capacity'] = (df_ts['annual_income'] / 12) * (1 - df_ts['debt_to_income_ratio']) * 0.3
df_ts['total_interest_cost'] = df_ts['loan_amount'] * df_ts['interest_rate']/ 100
df_ts['interest_to_principal_ratio'] = df_ts['total_interest_cost'] / df_ts['loan_amount']
df_ts['loan_burden_index'] = (df_ts['loan_amount'] * df_ts['interest_rate'] / 100) / df_ts['annual_income']
```

```python
# normalize numeric shit
from sklearn.preprocessing import StandardScaler

normalize_cols = [
    "annual_income",
    "debt_to_income_ratio",
    "credit_score",
    "loan_amount",
    "interest_rate",
    "loan_to_income_ratio",
    "disposable_income",
    "monthly_payment_capacity",
    "total_interest_cost",
    "interest_to_principal_ratio",
    "loan_burden_index"
]

scaler = StandardScaler()
df_tr[normalize_cols] = scaler.fit_transform(df_tr[normalize_cols])
df_ts[normalize_cols] = scaler.fit_transform(df_ts[normalize_cols])
```

```python
from helper_functions import plot_predictions, plot_decision_boundary
def plot(X, y, model= None):
        if model is not None:
            plt.figure(figsize=(16, 8))
            plt.subplot(1, 2, 1)
            plt.title("Train")
            plot_decision_boundary(model, X, y)
        else:
            plt.figure(figsize=(12,12))
            plt.scatter(X[:,0], X[:,1], c=y, cmap=plt.cm.RdYlBu)
```

```python
from sklearn.model_selection import train_test_split

X = df_tr.drop(columns=["loan_paid_back"])
y = df_tr["loan_paid_back"]

X_train, X_val, y_train, y_val = train_test_split(X,y,test_size=0.05,random_state=42)

len(X_train), len(X_val), len(y_train), len(X_val)
```

```python
from catboost import CatBoostClassifier
from sklearn.metrics import accuracy_score, roc_auc_score, classification_report


cb_model = CatBoostClassifier(
    iterations=600,
    learning_rate=0.1,
    depth=6,
    l2_leaf_reg=7.011376544533179,
    border_count=150,
    random_seed=42,
    verbose=100
)

cb_model.fit(X_train, y_train, eval_set=(X_val, y_val), use_best_model=True)
```

```python
preds = cb_model.predict(X_val)
preds_tr = cb_model.predict(X_train)

acc = accuracy_score(y_val, preds)
acc_tr = accuracy_score(y_train, preds_tr)

print("\nCatBoost Evaluation")
print(f"test set: {acc:.4f} | train set: {acc_tr:.4f}")
print(classification_report(y_val, preds))
```
