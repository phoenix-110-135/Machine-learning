```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_tr = pd.read_csv("data/train.csv").drop(columns = "id")
df_ts = pd.read_csv("data/test.csv").drop(columns = "id")

df_tr
```

```python
# map some shit with binary

df_ts["gender"] = df_ts["gender"].map({"Male": 1, "Female": 0, "Other": -1})
df_tr["gender"] = df_tr["gender"].map({"Male": 1, "Female": 0, "Other": -1})



order = ["No formal", "Highschool", "Graduate", "Postgraduate"]
df_tr["education_level"] = (
    pd.Categorical(df_tr["education_level"], categories=order, ordered=True)
    .codes
)
df_ts["education_level"] = (
    pd.Categorical(df_ts["education_level"], categories=order, ordered=True)
    .codes
)



order = ["Never", "Former", "Current"]
df_tr["smoking_status"] = (
    pd.Categorical(df_tr["smoking_status"], categories=order, ordered=True)
    .codes
)
df_ts["smoking_status"] = (
    pd.Categorical(df_ts["smoking_status"], categories=order, ordered=True)
    .codes
)



order = ["Student", "Unemployed", "Retired", "Employed"]
df_tr["employment_status"] = (
    pd.Categorical(df_tr["employment_status"], categories=order, ordered=True)
    .codes
)
df_ts["employment_status"] = (
    pd.Categorical(df_ts["employment_status"], categories=order, ordered=True)
    .codes
)



order = ["Low", "Lower-Middle", "Middle", "Upper-Middle", "High"]
df_tr["income_level"] = (
    pd.Categorical(df_tr["income_level"], categories=order, ordered=True)
    .codes
)
df_ts["income_level"] = (
    pd.Categorical(df_ts["income_level"], categories=order, ordered=True)
    .codes
)
```

```python
#get dummy of some categorical shit

dummies = pd.get_dummies(df_tr["ethnicity"], dtype=int)
df_tr = df_tr.drop(columns=["ethnicity"]).join(dummies)

dummies = pd.get_dummies(df_ts["ethnicity"], dtype=int)
df_ts = df_ts.drop(columns=["ethnicity"]).join(dummies)
df_ts
```

```python
# ===== Feature Engineering =====

# Blood pressure features
df_tr["pulse_pressure"] = df_tr["systolic_bp"] - df_tr["diastolic_bp"]
df_tr["mean_arterial_pressure"] = (
    df_tr["diastolic_bp"] + df_tr["pulse_pressure"] / 3
)

# Cholesterol ratios
df_tr["ldl_hdl_ratio"] = df_tr["ldl_cholesterol"] / df_tr["hdl_cholesterol"]
df_tr["total_hdl_ratio"] = df_tr["cholesterol_total"] / df_tr["hdl_cholesterol"]

# BMI-based features
df_tr["bmi_risk"] = pd.cut(
    df_tr["bmi"],
    bins=[0, 18.5, 25, 30, 100],
    labels=[0, 1, 2, 3]
).astype(int)

df_tr["central_obesity"] = (df_tr["waist_to_hip_ratio"] > 0.9).astype(int)

# Lifestyle balance
df_tr["activity_screen_ratio"] = (
    df_tr["physical_activity_minutes_per_week"] /
    (df_tr["screen_time_hours_per_day"] * 7 + 1)
)

df_tr["sleep_deviation"] = abs(df_tr["sleep_hours_per_day"] - 7)

# Risk aggregation
df_tr["metabolic_risk_score"] = (
    (df_tr["bmi"] > 30).astype(int) +
    (df_tr["triglycerides"] > 150).astype(int) +
    (df_tr["hdl_cholesterol"] < 40).astype(int) +
    (df_tr["systolic_bp"] > 130).astype(int)
)

df_tr["genetic_cardiac_risk"] = (
    df_tr["family_history_diabetes"] +
    df_tr["hypertension_history"] +
    df_tr["cardiovascular_history"]
)

# Age interactions
df_tr["age_squared"] = df_tr["age"] ** 2
df_tr["age_bmi_interaction"] = df_tr["age"] * df_tr["bmi"]
df_tr["age_bp_interaction"] = df_tr["age"] * df_tr["systolic_bp"]

# Lifestyle compression
df_tr["lifestyle_risk_index"] = (
    df_tr["alcohol_consumption_per_week"] +
    df_tr["screen_time_hours_per_day"] -
    (df_tr["physical_activity_minutes_per_week"] / 60)
)
```

```python
# ===== Feature Engineering (on test data) =====

# Blood pressure features
df_ts["pulse_pressure"] = df_ts["systolic_bp"] - df_ts["diastolic_bp"]
df_ts["mean_arterial_pressure"] = (
    df_ts["diastolic_bp"] + df_ts["pulse_pressure"] / 3
)

# Cholesterol ratios
df_ts["ldl_hdl_ratio"] = df_ts["ldl_cholesterol"] / df_ts["hdl_cholesterol"]
df_ts["total_hdl_ratio"] = df_ts["cholesterol_total"] / df_ts["hdl_cholesterol"]

# BMI-based features
df_ts["bmi_risk"] = pd.cut(
    df_ts["bmi"],
    bins=[0, 18.5, 25, 30, 100],
    labels=[0, 1, 2, 3]
).astype(int)

df_ts["central_obesity"] = (df_ts["waist_to_hip_ratio"] > 0.9).astype(int)

# Lifestyle balance
df_ts["activity_screen_ratio"] = (
    df_ts["physical_activity_minutes_per_week"] /
    (df_ts["screen_time_hours_per_day"] * 7 + 1)
)

df_ts["sleep_deviation"] = abs(df_ts["sleep_hours_per_day"] - 7)

# Risk aggregation
df_ts["metabolic_risk_score"] = (
    (df_ts["bmi"] > 30).astype(int) +
    (df_ts["triglycerides"] > 150).astype(int) +
    (df_ts["hdl_cholesterol"] < 40).astype(int) +
    (df_ts["systolic_bp"] > 130).astype(int)
)

df_ts["genetic_cardiac_risk"] = (
    df_ts["family_history_diabetes"] +
    df_ts["hypertension_history"] +
    df_ts["cardiovascular_history"]
)

# Age interactions
df_ts["age_squared"] = df_ts["age"] ** 2
df_ts["age_bmi_interaction"] = df_ts["age"] * df_ts["bmi"]
df_ts["age_bp_interaction"] = df_ts["age"] * df_ts["systolic_bp"]

# Lifestyle compression
df_ts["lifestyle_risk_index"] = (
    df_ts["alcohol_consumption_per_week"] +
    df_ts["screen_time_hours_per_day"] -
    (df_ts["physical_activity_minutes_per_week"] / 60)
)
```

```python
# normalize numeric shit
from sklearn.preprocessing import StandardScaler

normalize_cols = [
    "age",
    "physical_activity_minutes_per_week",
    "diet_score",
    "sleep_hours_per_day",
    "screen_time_hours_per_day",
    "bmi",
    "waist_to_hip_ratio",
    "systolic_bp",
    "diastolic_bp",
    "heart_rate",
    "cholesterol_total",
    "hdl_cholesterol",
    "triglycerides"
]

scaler = StandardScaler()
df_tr[normalize_cols] = scaler.fit_transform(df_tr[normalize_cols])
```

```python
import torch
from torch import nn
from sklearn.model_selection import train_test_split
from torch.utils.data import DataLoader, TensorDataset


X = df_tr.drop(columns=["diagnosed_diabetes"]).values
y = df_tr["diagnosed_diabetes"].values

X_train, X_val, y_train, y_val = train_test_split(X,y,test_size=0.1,random_state=42)

len(X_train), len(X_val), len(y_train), len(X_val)
```

```python
# Convert to tensors
X_train = torch.tensor(X_train, dtype=torch.float32)
X_val   = torch.tensor(X_val, dtype=torch.float32)
y_train = torch.tensor(y_train, dtype=torch.float32).unsqueeze(1)
y_val   = torch.tensor(y_val, dtype=torch.float32).unsqueeze(1)

# DataLoader
train_ds = TensorDataset(X_train, y_train)
val_ds   = TensorDataset(X_val, y_val)

train_loader = DataLoader(train_ds, batch_size=256, shuffle=True)
val_loader   = DataLoader(val_ds, batch_size=256)
```

```python
class DiabetesNN(nn.Module):
    def __init__(self, input_dim):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(input_dim, 32),
            nn.ReLU(),
            nn.BatchNorm1d(32),
            nn.Linear(32, 16),
            nn.ReLU(),
            nn.BatchNorm1d(16),
            nn.Linear(16, 1)   
        )

    def forward(self, x):
        return self.net(x)
```

```python
model = DiabetesNN(input_dim=X_train.shape[1])
```

```python
criterion = nn.BCEWithLogitsLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.05)
```

```python
import xgboost as xgb
from catboost import CatBoostClassifier
from sklearn.metrics import accuracy_score, roc_auc_score, classification_report

# ---------------------------
# XGBoost with best parameters
# ---------------------------
xgb_model = xgb.XGBClassifier(
    max_depth=8,
    learning_rate=0.10598584575917368,
    subsample=0.7039501207226441,
    colsample_bytree=0.5645876077433202,
    gamma=1.0416596272841665,
    min_child_weight=4,
    reg_alpha=3.2802242210265797,
    reg_lambda=1.222632035374115,
    n_estimators=109,
    objective="binary:logistic",
    eval_metric="auc",
    use_label_encoder=False,
    tree_method="hist",
    random_state=42
)

xgb_model.fit(X_train, y_train, eval_set=[(X_val, y_val)], verbose=False)

# Predictions & evaluation
preds_xgb = xgb_model.predict(X_val)
probs_xgb = xgb_model.predict_proba(X_val)[:,1]

acc_xgb = accuracy_score(y_val, preds_xgb)
auc_xgb = roc_auc_score(y_val, probs_xgb)

print("XGBoost Evaluation")
print(f"Accuracy: {acc_xgb:.4f} | ROC-AUC: {auc_xgb:.4f}")
print(classification_report(y_val, preds_xgb))


# ---------------------------
# CatBoost with best parameters
# ---------------------------
cb_model = CatBoostClassifier(
    iterations=490,
    learning_rate=0.2704733846104388,
    depth=6,
    l2_leaf_reg=7.011376544533179,
    border_count=139,
    random_seed=42,
    verbose=100
)

cb_model.fit(X_train, y_train, eval_set=(X_val, y_val), use_best_model=True)

# Predictions & evaluation
preds_cb = cb_model.predict(X_val)
probs_cb = cb_model.predict_proba(X_val)[:,1]

acc_cb = accuracy_score(y_val, preds_cb)
auc_cb = roc_auc_score(y_val, probs_cb)

print("\nCatBoost Evaluation")
print(f"Accuracy: {acc_cb:.4f} | ROC-AUC: {auc_cb:.4f}")
print(classification_report(y_val, preds_cb))


# ---------------------------
# Compare
# ---------------------------
if auc_cb > auc_xgb:
    print("\n✅ CatBoost is the best model based on ROC-AUC!")
else:
    print("\n✅ XGBoost is the best model based on ROC-AUC!")
```
