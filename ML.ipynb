```python
from sklearn.neighbors   import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble    import RandomForestClassifier
```

```python
import numpy   as np 
import pandas  as pd 
import seaborn as sns
```

```python
from  sklearn.model_selection import  train_test_split
from  sklearn.metrics         import  classification_report
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
iris = pd.read_csv('File/iris.csv') 
iris
```

```python
X = iris.drop(['species'],axis=1)
X
```

```python
y = iris['species']
y
```

```python
g = sns.FacetGrid(iris, hue="species")
g.map(sns.scatterplot, "petallength", "petalwidth", alpha=.7)
g.add_legend(loc=1)
```

```python
sns.boxplot(x="species", y="petallength", data=iris);
```

```python
X_train, X_test, y_train, y_test = train_test_split( X, y, stratify=y, test_size=0.3, random_state=42)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test .shape
```

```python
knn = KNeighborsClassifier(n_neighbors=3)   # n_neighbors=3
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)
print(classification_report(y_test, y_pred))  
```

```python
nb = GaussianNB()

nb.fit(X_train, y_train)
y_pred = nb.predict(X_test)
print(classification_report(y_test, y_pred))  
```

```python
rf = RandomForestClassifier(n_estimators=200,random_state=42,max_depth=3)  # n_estimators=100, random_state=123
rf.fit(X_train, y_train)
y_pred = rf.predict(X_test)
print(classification_report(y_test, y_pred))  
```
